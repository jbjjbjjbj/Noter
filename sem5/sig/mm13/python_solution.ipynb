```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataorig = loadmat("signal.mat")
dataorig = dataorig["s"]
np.savetxt("signal.txt", dataorig)
```

## Opgave 1

```python
# Genereret med ./a.out -i signal.txt -f 1024 -o output1.txt
fft = np.loadtxt("output1.txt", dtype=complex)
fft
```

```python
step = 8000/len(fft)
x = np.arange(len(fft)) * step
plt.plot(x, np.abs(fft))
```

## Opgave 2

```python
step = 8000/len(fft)
x = np.arange(len(fft)) * step
plt.xlim(0, 100)
plt.plot(x, np.abs(fft))
```
