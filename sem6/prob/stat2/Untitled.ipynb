## Part A

As $\alpha \cdot S$ is a constant, $y_i$ will also have a gaussian distribution with $\mu = \alpha \cdot S$ and $\sigma^2$.
Therefore we can use maximum likelyhood estimation for a gaussian random variable.

$$
\hat{\mu} = \frac 1 N \sum_{i=1}^N x_i
$$

From this we can isolate for $d$.

$$
\mu = \alpha \cdot S = \frac {0.5} d \cdot S\\
d = \frac S {\mu \cdot 2} \\
\hat{d} = \frac {S \cdot N} {2 \cdot \sum_{i=1}^N y_i}
$$

First we define the likelyhood function.

$$
f(x_1, x_2 | \mu) = P(X_1 = x_1, X_2 = x_2 | \mu)
$$

Here $X_1 \sim \mathcal{N}(\mu, \sigma_1^2)$ and $X_2 \sim \mathcal{N}(\mu, \sigma_2^2)$.
Because $X_1$ and $X_2$ are independend, they can be multiplied together.

$$
f(x_1, x_2 | \mu) = P(X_1 = x_1 | \mu) \cdot P(X_2 = x_2 | \mu) \\
= \frac 1 {0.5 \cdot \sqrt{2 \pi}} \cdot \exp\left(- \frac {(x_1 - \mu)^2} {2 \cdot 0.5}\right) \cdot \frac 1 {2 \cdot \sqrt{2 \pi}} \cdot \exp\left(- \frac {(x_2 - \mu)^2} {2 \cdot 2}\right) \\
= \frac 1 {2 \pi} \cdot \exp\left(- (x_1 - \mu)^2 - \frac 1 4 (x_2 - \mu)^2\right) \\
\log f(x_1, x_2 | \mu) = - \log (2 \pi) - (x_1 - \mu)^2 - \frac 1 4 (x_2 - \mu)^2
$$

Nu kan vi tage differentialet, og lse for 0.

$$
